```python
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt

# Create a 2D NumPy array with random values
array = np.random.randint(10, size=(5, 5))

# Print the array as a table
print(tabulate(array, tablefmt="grid"))

# Plot the array as a heatmap
plt.imshow(array, cmap="plasma", interpolation="nearest")
plt.show()
```

```python
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt

# Create a 3D NumPy array with random values
array = np.random.randint(10, size=(3, 3, 3))

# Flatten the array into a 2D array
flat_array = array.flatten()

# Print the array as a table
print(tabulate(flat_array, tablefmt="grid"))

# Plot the array as a heatmap
plt.imshow(flat_array, cmap="plasma", interpolation="nearest")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# Create a 3D array
X, Y, Z = axes3d.get_test_data(0.5)

# Plot the 3D array
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

```python
import plotly.express as px
import numpy as np

# Create a 3D array
X, Y, Z = np.mgrid[-5:5:64j, -5:5:64j, -5:5:64j]
data = np.sin(X**2 + Y**2 + Z**2)

# Plot the 3D array
fig = px.scatter_3d(x=X.flatten(), y=Y.flatten(), z=Z.flatten(), color=data.flatten())
fig.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and an axis
fig, ax = plt.subplots()

# Plot some data
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Add a grid to the plot
ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Create a figure and an axis
fig, ax = plt.subplots()

# Plot some data
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Set the major and minor grid size
# ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(0.5))
# ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))

# Add a grid to the plot
ax.grid(which='major', color='k', linestyle='-', linewidth=1)
ax.grid(which='minor', color='k', linestyle=':', linewidth=1)

# Show the plot
plt.show()
```

```python
#Import tkinter library
from tkinter import *
#Create an instance of Tkinter frame or window
win= Tk()
#Set the geometry of tkinter frame
win.geometry("750x250")
def callback():
   Label(win, text="Hello World!", font=('Century 20 bold')).pack(pady=4)
#Create a Label and a Button widget
btn=Button(win, text="Press Enter", command= callback)
btn.pack(ipadx=10)
win.bind('<Return>',lambda event:callback())
win.mainloop()
```
